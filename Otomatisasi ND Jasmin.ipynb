```python
import os
import time
import base64
import ipywidgets as widgets
from IPython.display import display, HTML
from docxtpl import DocxTemplate

# 1. PENGATURAN TEMPLATE & FOLDER TUJUAN
TEMPLATE_PATH = "Master ND.docx"
FOLDER_TUJUAN = r"D:\jasmin"

# ==========================================
# 🎨 CONVERT LOGO BPK (BASE64)
# ==========================================
def get_base64_image(image_path):
    try:
        if os.path.exists(image_path):
            with open(image_path, "rb") as image_file:
                return base64.b64encode(image_file.read()).decode('utf-8')
    except Exception as e:
        print(f"Gagal membaca Base64: {e}")
    return ""

encoded_logo = get_base64_image("R.png")
if encoded_logo:
    logo_src = f"data:image/png;base64,{encoded_logo}"
else:
    logo_src = "https://upload.wikimedia.org/wikipedia/commons/thumb/8/87/Logo_BPK-RI.png/240px-Logo_BPK-RI.png"

# ==========================================
# 🎨 STYLING UTAMA (FORM LEBAR & WARNA TETAP)
# ==========================================
global_style = """
<style>
    body, .jp-Notebook, .vuetify-app, .jupyter-widgets-view {
        background-image: linear-gradient(rgba(0, 51, 102, 0.75), rgba(0, 51, 102, 0.85)), 
                          url('https://www.bpk.go.id/assets/images/slideshow/slide_1.jpg') !important;
        background-size: cover !important;
        background-position: center center !important;
        background-attachment: fixed !important;
        font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
    }

    /* Kartu Form Utama - Dilebarkan ke 1200px agar tidak ada horizontal scroll */
    .widget-vbox.widget-box-custom {
        background: linear-gradient(135deg, rgba(255, 255, 255, 0.98) 60%, rgba(255, 225, 100, 0.15) 100%) !important;
        padding: 35px 45px 50px 45px !important;
        border-radius: 12px !important;
        box-shadow: 0px 15px 35px rgba(0, 0, 0, 0.5) !important;
        border-top: 6px solid #FFD700 !important; 
        border-bottom: 4px solid #003366 !important;
        max-width: 1200px;
        width: 95% !important;
        margin: 30px auto !important;
        position: relative !important;
    }
    
    .widget-vbox.widget-box-custom .custom-title-main { color: #003366 !important; }
    .widget-vbox.widget-box-custom .custom-title-sub { color: #666666 !important; }
    .widget-vbox.widget-box-custom .clock-container { background: rgba(0, 0, 0, 0.05); color: #555; }
    .widget-vbox.widget-box-custom .footer-text-left { color: #888 !important; }
    .widget-vbox.widget-box-custom .footer-text-right { color: #003366 !important; }

    /* Memastikan tata letak baris widget menggunakan ruang penuh */
    .widget-hbox {
        width: 100% !important;
        justify-content: space-between !important;
        margin-bottom: 10px !important;
    }
    
    /* Lebar dinamis elemen input di dalam baris */
    .widget-hbox .widget-text, .widget-hbox .widget-textarea {
        flex: 1 1 auto !important;
    }

    /* Jam Digital Kanan Atas */
    .clock-container {
        position: absolute;
        top: 20px;
        right: 25px;
        font-family: 'Courier New', Courier, monospace;
        font-size: 20px;
        font-weight: bold;
        padding: 6px 12px;
        border-radius: 6px;
        border: 1px solid rgba(128,128,128,0.2);
        z-index: 999;
    }
</style>

<script>
    // System Realtime Jam Laptop Lokal
    function startLiveClock() {
        setInterval(function() {
            var now = new Date();
            var days = ['Minggu', 'Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu'];
            var dayName = days[now.getDay()];
            var date = String(now.getDate()).padStart(2, '0');
            var month = String(now.getMonth() + 1).padStart(2, '0');
            var year = now.getFullYear();
            var hours = String(now.getHours()).padStart(2, '0');
            var minutes = String(now.getMinutes()).padStart(2, '0');
            var seconds = String(now.getSeconds()).padStart(2, '0');
            
            var clockElements = document.getElementsByClassName('clock-display');
            var clockStr = dayName + ", " + date + "/" + month + "/" + year + " - " + hours + ":" + minutes + ":" + seconds;
            
            for (var i = 0; i < clockElements.length; i++) {
                clockElements[i].innerText = clockStr;
            }
        }, 1000);
    }
    setTimeout(startLiveClock, 500);
</script>
"""
display(HTML(global_style))

# Header Title Form + Widget Wadah Jam Digital (Kanan Atas)
html_title = widgets.HTML(value=f"""
<div class="clock-container"><span class="clock-display">Memuat Jam...</span></div>
<div style='text-align: center; margin-bottom: 25px;'>
    <img src='{logo_src}' width='120' style='margin-bottom: 12px;' onerror="this.src='https://upload.wikimedia.org/wikipedia/commons/thumb/8/87/Logo_BPK-RI.png/240px-Logo_BPK-RI.png'"><br>
    <h2 class="custom-title-main" style='margin: 0; font-family: sans-serif; font-size: 26px; font-weight: 800; letter-spacing: 1px;'>
        Aplikasi Nota Dinas Direktorat IV.B.2
    </h2>
    <p class="custom-title-sub" style='font-style: italic; margin-top: 6px; font-size: 13px;'>
        "Integritas, Independensi dan Profesionalisme"
    </p>
    <hr style='border: 0; border-top: 1px solid #ddd; margin-top: 15px; width: 80%;'>
</div>
""")

# Footer Teks & Penulis
html_author = widgets.HTML(value="""
<div style='text-align: center; margin-top: 5px; color: #777; font-size: 8px; font-style: italic; font-weight: 500;'>
    Airul Nurpahmi
</div>
""")

html_footer_labels = widgets.HTML(value="""
<div style='margin-top: 35px; border-top: 1px dashed #ddd; padding-top: 15px; font-size: 12px; font-weight: bold;'>
    <div style='float: left; font-style: italic;' class='footer-text-left'>
        📌 "Integritas Bukan Hanya Diatas Kertas"
    </div>
    <div style='float: right; letter-spacing: 0.5px;' class='footer-text-right'>
        🏢 Direktorat IV.B
    </div>
    <div style='clear: both;'></div>
</div>
""")

# 2. KOMPONEN FORM INPUT (Dengan Layout Flexible & Lebar Terdistribusi Otomatis)
layout_4_col = widgets.Layout(width='24%', margin='0px 0.5%')
layout_3_col = widgets.Layout(width='32%', margin='0px 0.5%')
layout_2_col = widgets.Layout(width='49%', margin='0px 0.5%')

txt_no = widgets.Text(description='No Surat (no):', placeholder='123', layout=layout_4_col)
txt_kode = widgets.Text(description='Kode:', placeholder='PMT.01', layout=layout_4_col)
txt_bulan = widgets.Text(description='Bulan Surat:', placeholder='08', layout=layout_4_col)
txt_tahun = widgets.Text(description='Tahun Surat:', placeholder='2026', layout=layout_4_col)

txt_lampiran = widgets.Text(description='Lampiran:', placeholder='Contoh: Satu Berkas', layout=layout_2_col)
txt_perihal = widgets.Text(description='Perihal:', placeholder='Ketik perihal di sini...', layout=layout_2_col)

txt_tgl_ctk = widgets.Text(description='Tanggal:', placeholder='08', layout=layout_3_col)
txt_bln_ctk = widgets.Text(description='Bulan:', placeholder='Contoh: Juli', layout=layout_3_col)
txt_thn_ctk = widgets.Text(description='Tahun:', placeholder='2026', layout=layout_3_col)

txt_isi = widgets.Textarea(
    description='Isi Surat:',
    placeholder='Ketik isi surat secara lengkap di sini...',
    layout=widgets.Layout(width='99%', height='180px', margin='0px 0.5%')
)

btn_proses = widgets.Button(
    description='Buat & Simpan Word',
    button_style='info', 
    icon='file-word',
    layout=widgets.Layout(width='25%', height='42px')
)

out_message = widgets.Output()

# 3. FUNGSI TOMBOL PROSES SURAT
def on_button_clicked(b):
    b.description = "Sedang Memproses Dokumen..."
    b.disabled = True
    with out_message:
        out_message.clear_output()
        loading_bar = widgets.IntProgress(value=0, min=0, max=10, step=1, description='Progress:', bar_style='success', layout=widgets.Layout(width='50%'))
        display(loading_bar)
        for i in range(1, 11):
            time.sleep(0.05)
            loading_bar.value = i
            
        if not os.path.exists(TEMPLATE_PATH):
            print(f"❌ File '{TEMPLATE_PATH}' tidak ditemukan!")
            b.description = "Buat & Simpan Word"; b.disabled = False; return
        try:
            if not os.path.exists(FOLDER_TUJUAN): os.makedirs(FOLDER_TUJUAN)
            doc = DocxTemplate(TEMPLATE_PATH)
            context = {
                'no': txt_no.value, 'kode': txt_kode.value, 'bulan': txt_bulan.value, 'tahun': txt_tahun.value,
                'lampiran': txt_lampiran.value, 'perihal': txt_perihal.value, 'tanggal': txt_tgl_ctk.value,
                'bulan_cetak': txt_bln_ctk.value, 'Tahun': txt_thn_ctk.value, 'isisurat': txt_isi.value
            }
            doc.render(context)
            nama_file = f"Nota_Dinas_{txt_no.value}.docx"
            doc.save(os.path.join(FOLDER_TUJUAN, nama_file))
            print(f"✔️ Berhasil disimpan di: {FOLDER_TUJUAN}")
        except Exception as e:
            print(f"❌ Terjadi kesalahan: {str(e)}")
            
    b.description = "Buat & Simpan Word"; b.disabled = False

btn_proses.on_click(on_button_clicked)

# 4. OUTPUT CONTAINER FORM UTAMA (TATA LETAK RESPONSIVITAS HORIZONTAL)
row_1 = widgets.HBox([txt_no, txt_kode, txt_bulan, txt_tahun])
row_2 = widgets.HBox([txt_lampiran, txt_perihal])
row_3 = widgets.HBox([txt_tgl_ctk, txt_bln_ctk, txt_thn_ctk])
row_4 = widgets.HBox([txt_isi])

# Terapkan class horizontal agar rapi dan melebar sempurna
row_1.add_class("widget-hbox")
row_2.add_class("widget-hbox")
row_3.add_class("widget-hbox")
row_4.add_class("widget-hbox")

form_container = widgets.VBox([
    html_title,
    row_1,
    row_2,
    row_3,
    row_4,
    widgets.Box([btn_proses], layout=widgets.Layout(justify_content='center', margin='20px 0px')),
    html_author,
    out_message,
    html_footer_labels
])

form_container.add_class("widget-vbox")
form_container.add_class("widget-box-custom")
display(form_container)
```
